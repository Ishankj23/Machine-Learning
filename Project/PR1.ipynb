```python
import pandas as pd
df=pd.read_csv('training_set.csv')
df.head()
```

```python
x=df.drop(columns=['Id', 'SalePrice'])
y=df[['SalePrice']]
```

```python
cat=list(x.columns[x.dtypes=='object'])
con=list(x.columns[x.dtypes!='object'])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
con_pipe1=Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                          ('scaler', StandardScaler())])
```

```python
cat_pipe1=Pipeline(steps=[('impute', SimpleImputer(strategy='constant', fill_value='Not_Avail')),
                          ('ordinal', OrdinalEncoder())])
```

```python
pre1=ColumnTransformer([('con', con_pipe1, con),
                        ('cat', cat_pipe1, cat)]).set_output(transform='pandas')
```

```python
x_pre = pre1.fit_transform(x)
```

```python
x_pre
```

```python
x_pre.isna().sum()
```

```python
cat_sel = list(x_pre.columns[x_pre.dtypes=='object'])
con_sel = list(x_pre.columns[x_pre.dtypes!='object'])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
con_pipe2 = Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),
                            ('scaler', StandardScaler())])
```

```python
cat_pipe2 = Pipeline(steps=[('impute', SimpleImputer(strategy='constant', fill_value='Not_Avail')),
                            ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
```

```python
pre2 = ColumnTransformer([('con', con_pipe2, con_sel),
                          ('cat', cat_pipe2, cat_sel)]).set_output(transform='pandas')
```

```python
pre2
```

```python
x_pre2 = pre2.fit_transform(x_pre)
x_pre2.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
lr = LinearRegression()
sel = SequentialFeatureSelector(estimator=lr, n_features_to_select='auto' , direction='forward')
sel.fit_transform(x_pre2, y)
sel_cols = sel.get_feature_names_out()
sel_cols
```

```python
org_col=[]
for i in sel_cols:
    s=i.split('__', '__')[1, 3]
    org_col.append(s)
```

```python
org_col
```

```python
x_sel=x[org_col]
x_sel.head()
```

```python
cat_sel = list(x_sel.columns[x_sel.dtypes=='object'])
con_sel = list(x_sel.columns[x_sel.dtypes!='object'])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
con_pipe2 = Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),
                            ('scaler', StandardScaler())])
```

```python
cat_pipe2 = Pipeline(steps=[('impute', SimpleImputer(strategy='constant')),
                            ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
```

```python
pre2 = ColumnTransformer([('con', con_pipe2, con_sel),
                          ('cat', cat_pipe2, cat_sel)]).set_output(transform='pandas')
```

```python
pre2
```

```python
x_sel_pre = pre2.fit_transform(x_sel)
x_sel_pre.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x_sel_pre, y, test_size=0.2, random_state=60)
```
