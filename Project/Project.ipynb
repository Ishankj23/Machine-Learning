```python
x = df.drop(columns=['Id', 'SalePrice'])
y = df[['SalePrice']]
```

```python
x.head()
```

```python
y.head()
```

```python
cat = list(x.columns[x.dtypes=='object'])
cat
```

```python
con = list(x.columns[x.dtypes!='object'])
con
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
con_pipe1 = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())])
```

```python
cat_pipe1 = Pipeline(steps=[('impute', SimpleImputer(strategy='constant', fill_value='Not_Avail')),
                            ('ordinal', OrdinalEncoder())])
```

```python
pre1 = ColumnTransformer([('con', con_pipe1, con),
                          ('cat', cat_pipe1, cat)]).set_output(transform='pandas')
```

```python
x_pre = pre1.fit_transform(x)
```

```python
x_pre.head()
```

```python
x_pre.isna().sum()
```

```python
x_pre.to_csv('training preprocessed.csv', index=False)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
lr = LinearRegression()
sel = SequentialFeatureSelector(estimator=lr, n_features_to_select='auto' , direction='backward')
sel.fit_transform(x_pre, y)
sel_cols = sel.get_feature_names_out()
sel_cols
```

```python
len(sel_cols)
```

```python
org_col=[]
for i in sel_cols:
    s=i.split('__')[1]
    org_col.append(s)
```

```python
org_col
```

```python
x_sel=x[org_col]
x_sel.head()
```

```python
cat_sel = list(x_sel.columns[x_sel.dtypes=='object'])
con_sel = list(x_sel.columns[x_sel.dtypes!='object'])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
con_pipe2 = Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),
                            ('scaler', StandardScaler())])
```

```python
cat_pipe2 = Pipeline(steps=[('impute', SimpleImputer(strategy='constant', fill_value='Not_Avail')),
                            ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
```

```python
pre2 = ColumnTransformer([('con', con_pipe2, con_sel),
                          ('cat', cat_pipe2, cat_sel)]).set_output(transform='pandas')
```

```python
pre2
```

```python
x_sel_pre = pre2.fit_transform(x_sel)
x_sel_pre.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x_sel_pre, y, test_size=0.2, random_state=21)
```

```python
xtest.head()
```

```python
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

```python
from sklearn.linear_model import Lasso 
model2 = Lasso(alpha=1)
model2.fit(xtrain, ytrain)
```

```python
model2.score(xtrain, ytrain)
```

```python
model2.score(xtest, ytest)
```

```python
import numpy as np 
params = {'alpha': np.arange(start=1, stop=1000, step=1)}
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
ls = Lasso()
gscv1 = GridSearchCV(estimator=ls, param_grid=params, cv=5, scoring='neg_mean_squared_error')
gscv1.fit(xtrain, ytrain)
```

```python

gscv1.best_params_
```

```python
gscv1.best_score_
```

```python

best_lasso = gscv1.best_estimator_
best_lasso
```

```python

best_lasso.score(xtrain, ytrain)
```

```python

best_lasso.score(xtest, ytest)
```

```python
def evaluate_model(model, x, y):
    ypred = model.predict(x)
    from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
    from sklearn.metrics import mean_absolute_percentage_error
    mse = mean_squared_error(y, ypred)
    rmse = mse**(1/2)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)
    print(f'Mean Squared Error : {mse:.2f}')
    print(f'Root Mean Squared Error : {rmse:.2f}')
    print(f'Mean Absolute Error : {mae:.2f}')
    print(f'Mean Absolute Percentage Error : {mape:.4f}')
    print(f'R2 Score : {r2:.4f}')
```

```python
evaluate_model(model2, xtrain, ytrain)
```

```python
evaluate_model(model2, xtest, ytest)
```

```python
from sklearn.model_selection import cross_val_score
scores_lasso = cross_val_score(best_lasso, xtrain, ytrain, cv=5, scoring='r2')
scores_lasso
```
