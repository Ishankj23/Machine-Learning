```python
import pandas as pd
df=pd.read_csv('/content/Mall_Customers.csv')
df.head()
```

```python
x=df[['Age', 'Income', 'Spending']]
x.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe=Pipeline([('impute', SimpleImputer(strategy='median')),
                   ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
x_pre=num_pipe.fit_transform(x)
x_pre.head()
```

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
```

```python
visualizer = KElbowVisualizer(KMeans(),
                              k=(2, 15),
                              timings=False)
visualizer.fit(x_pre)
visualizer.show()
```

```python
model1=KMeans(n_clusters=6)
lab1 = model1.fit_predict(x_pre)
```

```python
lab1
```

```python
x2=x.copy()
```

```python
x2
```

```python
x2['Label1']=lab1
x2
```

```python
x2[x2['Label1']==0]
```

```python
for i in range(6):
  print(f'Cluster number:{i}\n')
  display(x2[x2['Label1']==i])
  print('\n====================================\n')
```

```python
import plotly.express as px
fig1=px.scatter_3d(data_frame=x2,
                   x='Age',
                   y='Income',
                   z='Spending',
                   color='Label1')
fig1.show()
```

# 2.Agglomerative Clustering (Hierarchial Clustering)

```python
x_pre
```

```python
from sklearn.cluster import AgglomerativeClustering
model2= AgglomerativeClustering(n_clusters=6)
lab2=model2.fit_predict(x_pre)
```

```python
lab2
```

```python
x3=x.copy()
```

```python
x3
```

```python
x3['label2']=lab2
```

```python
x3
```

```python
for i in range(6):
  print(f'Cluster number : {i}\n')
  display(x3[x3['label2']==i])
  print('\n============================================\n')
```

```python
fig2 = px.scatter_3d(data_frame=x3,
                     x='Age',
                     y='Income',
                     z='Spending',
                     color='label2')
fig2.show()
```
